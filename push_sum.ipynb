# Push-sum algorithm

A gossip-based data aggregation algorithm for SUM and AVERAGE

```python
import networkx as nx
import numpy as np
from random import sample
```

```python
graph = nx.fast_gnp_random_graph(10, 0.5, seed=0, directed=True)
nx.draw(graph, with_labels=True)
```

# AVERAGE

```python
G = graph.copy()
for i in G.nodes:
    G.nodes[i]['s'] = {0: 5}
    G.nodes[i]['w'] = {0: 1}
# G.nodes[0]['w'][0] = 1

tf = 100
for t in range(1, tf):
    comm = dict()
    for i in G.nodes:
        G.nodes[i]['s'][t] = 0.5 * G.nodes[i]['s'][t-1]
        G.nodes[i]['w'][t] = 0.5 * G.nodes[i]['w'][t-1]
    for i in G.nodes:
        j = sample(list(G.neighbors(i)), 1)[0]
        # comm[i] = j
        G.nodes[j]['s'][t] += 0.5 * G.nodes[i]['s'][t-1]
        G.nodes[j]['w'][t] += 0.5 * G.nodes[i]['w'][t-1]
    # print(comm)
```

```python
for t in range(tf):
    agg = dict()
    for i in G.nodes:
        agg[i] = G.nodes[i]['w'][t] if G.nodes[i]['w'][t] > 0 else 0
    print(np.average(list(agg.values())))
```
